This doesn't really didn't hit what I wanted it to so I decided to switch to density plots instead

```python
fig, ax = plt.subplots()
sns.kdeplot(df3['percFemale'], ax=ax, label='Female')
sns.kdeplot(df3['percMale'], ax=ax,label='Male')
plt.legend()
```

What this does is plots the density of percentage so it shows that men were more often at a higher percentage of death meaning that they were more likely to die from the virus.
